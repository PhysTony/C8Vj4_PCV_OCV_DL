```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
def gray_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

**TASK: Open and display the giraffes.jpg image that is located in the DATA folder.**

```python
img_bgr = cv2.imread('../DATA/giraffes.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_gray = cv2.imread('../DATA/giraffes.jpg',0)
```

```python
# Verifiyng if an image was readded.
type(img_rgb)
```

```python
display_img(img_rgb)
```

**TASK:Apply a binary threshold onto the image.**

We will choose a threshold equals to middel point range of $[0,255]$.

```python
ret,thresh = cv2.threshold(img_gray,127,255,cv2.THRESH_BINARY)
```

```python
gray_img(thresh)
```

Now, a threshold equals to 150.

```python
ret,thresh = cv2.threshold(img_gray,150,255,cv2.THRESH_BINARY)
```

```python
gray_img(thresh)
```

Finally, a threshold equals to 100.

```python
ret,thresh = cv2.threshold(img_gray,100,255,cv2.THRESH_BINARY)
```

```python
gray_img(thresh)
```

**TASK: Open the giaraffes.jpg file from the DATA folder and convert its colorspace to  HSV and display the image.**

```python
img_bgr = cv2.imread('../DATA/giraffes.jpg')
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
```

```python
display_img(img_hsv)
```

**TASK: Create a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the giraffer image (displayed in normal RGB)**

```python
kernel = np.ones((4,4))/10
```

```python
kernel
```

```python
dst = cv2.filter2D(img_rgb,-1,kernel)
display_img(dst)
```

**TASK: Create a Horizontal Sobel Filter (sobelx from our lecture) with a kernel size of 5 to the grayscale version of the giaraffes image and then display the resulting gradient filtered version of the image.**

```python
sobelx = cv2.Sobel(img_gray,cv2.CV_64F,1,0,ksize=5)
```

```python
gray_img(sobelx)
```

**TASK: Plot the color histograms for the RED, BLUE, and GREEN channel of the giaraffe image. Pay careful attention to the ordering of the channels.** The image must be in **BGR**

```python
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img_bgr],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('Giraffe Image')
plt.show()
```
